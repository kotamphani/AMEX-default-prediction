```python
import dask
import dask.dataframe as dd
import pandas as pd
import numpy as np
```

```python
train_data = dd.read_csv('train_data.csv')
```

```python
train_data.head()
```

```python
train_data = train_data.groupby('customer_ID').apply(lambda x: x.sample(n=1, random_state = np.random.randint(1000)))
train_data = train_data.reset_index(drop = True)
```

```python
train_data.dtypes
```

```python
train_data_modified = train_data.compute()
train_data_modified.head()
```

```python
train_data_modified['S_2'] = pd.to_datetime(train_data_modified['S_2'])
train_data_modified['month_year'] = train_data_modified['S_2'].dt.strftime('%Y-%m')
```

```python
print(train_data_modified.shape)
train_data_modified['month_year'].value_counts()
```

```python
print(train_data_modified.head())
train_data_modified.shape
```

```python
train_data_modified.to_csv('final_data.csv', index = False)
```
